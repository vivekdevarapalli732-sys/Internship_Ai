```python
name = input("enter your name")
var = input("var to break:")
for i in name:
    if i == var:
       break
    print(i,end="")
print("\n Loop ended")
```

```python
name = input("enter your name")
var = input("var to break:")
for i in name:
    if i == var:
       continue
    print(i,end="")
print("\n Loop ended")
```

```python
name = input("enter your name")
var = input("var to break:")
for i in name:
    if i == var:
       pass
    print(i,end="")
print("\n Loop ended")
```
