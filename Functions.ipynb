```python
# function without parameters
def sample():
  print("Good Morning")
sample()
```

```python
# function with parameter
def sample(name): # function declearation
  print("My name is:",name) #block of code
sample("john") # function call
```

```python
#function with return type
def add(a,b):
  return a+b
print(add(2,3))
```

```python
# variales
# local variables
X = 10

def f1():
    print("X value from f1:", X)

def f2():
    print("X value from f2:", X)

f1()
f2()
print("X outside the function:", X)
```

```python
# Global variable
X = 10
def f1():
  print("X value from f1:",X)
def f2():
  print("X value from f2:",X)
f1()
f2()
print("X Value:",X)
```

```python
#Arguments
def add(a,b):  # formal arguments
  print(a+b)
add(20,60)     # actual arguments
```

```python
# position arguments
def Sample(a,b):
  print(a,b)
Sample(2,3)
Sample(3,2)
```

```python
def name(first_name,last_name):  # position arguments
  print(first_name,last_name)
name("Swanitha","Bandaru")
```

```python
# Keyword arguments
def sample(a,b):
  print(a,b)
sample(a=2,b=3)
sample(b=3,a=2)
```

```python
# keyword arguments
def greeting(msg, name):
    print(msg, name)

greeting("Good Morning", "Swanitha")
```

```python

#default arugument
def student(name, dept="CSD"):
    print("My name:", name, "\nDepartment:", dept)

student(name="Havi")
student(name="Havi", dept="CSE")
```

```python
def student_get_marks(results = "pass"):
    print("Result:", results)

student_get_marks()
student_get_marks("A+")
```

```python
# variable length arguments
def name(*n):
    print(n)

name(1)
name(1, 2)
name(1, 2, 3)
```

```python
# calculator.py

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

print("Addition =", add(10, 5))
print("Subtraction =", sub(10, 5))
print("Multiplication =", mul(10, 5))
print("Division =", div(10, 5))
```
