```python
l = [1,2,3,4,5]
print(l)

# append
l.append(6)
print(l)

# insert
l.insert(0,0)
print(l)

l1 = [7,8,9]

# extend
l.extend(l1)
print(l)

# remove
l.remove(4)
print(l)

# pop
l.pop(0)
print(l)

# index
print(l.index(3))

# count
print(l.count(5))

# sort
l.sort()
print(l)

# reverse
l.reverse()
print(l)

# copy
l2 = l.copy()
print(l2)

# len
length = len(l)
print("length:", length)

# max
maximum = max(l)
print("maximum:", maximum)

# min
minimum = min(l)
print("minimum:", minimum)

```

```python
#slicing
l = [1,2,3,4,5]
print(l[3:0:-1])
```

```python
#indexing
print(l[0])
print(l[1])
```
