# Copies And Viewing

```python
a = np.arange(10)
b = a[::2]
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

```python
b[0] = 10
print(a)
print(b)
```

```python
a = np.arange(10)
b = a[::2].copy()  # force a copy
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

```python
b[0] = 10
print(a)
print(b)
```
